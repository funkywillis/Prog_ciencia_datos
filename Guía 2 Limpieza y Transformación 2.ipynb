### Al analizar esta base, notamos que algunas filas de la variable "NIT" incluyeron los registros de las demás columnas. Esta columna debería tener unicamente los 9 caracteres del codigo NIT. Por lo tanto, esta es la primera limpieza que debe realizarse. Empezamos por consultar el tamaño del dataframe (filas y columnas) y luego conocer el tipo de dato que actualmente posee cada variable.

```python
df_empresas.shape
```

```python
df_empresas.dtypes
```

```python
df_empresas['NIT'] = df_empresas['NIT'].astype(str) #Se inicia con la transformación del tipo de dato de Object a String en NIT
```

```python
nit_invalid_length = df_empresas[df_empresas['NIT'].str.len() != 9] #Luego debe analizarse si la cadena tiene nueva digitos.
```

```python
num_invalid_nit = nit_invalid_length.shape[0]
```

```python
df_empresas.dtypes
```

```python
# Identificar filas problemáticas (donde NIT tiene más de 9 caracteres)
# Estas filas contienen observaciones erróneas que se han mezclado con los datos de otras columnas
problematic_rows = df_empresas[df_empresas['NIT'].str.len() > 9]
```

```python
import re

# Usar expresión regular para dividir solo en comas seguidas de texto/números y no seguidas de espacios
# Esto se puede hacer utilizando expresiones regulares para asegurarse de que las comas seguidas de un texto o número sean las que se utilicen para la separación.

split_data = problematic_rows['NIT'].str.split(r',(?=\S)', expand=True)
```

```python
split_data.head() #Ahora se verifica cuántas columnas se han generado para asegurar de que la división fue exitosa.
```

#### Ahora que tenemos las observaciones divididas correctamente, reasignaremos estos valores a las columnas originales de df_empresas:

```python
# Primero, es necesario redistribuir los valores en las columnas correctas
# Luego, hay que asegurarse de que las columnas de split_data coinciden con las columnas del DataFrame df_empresas

column_names = df_empresas.columns

# Reasignar los valores de split_data a las columnas del dataframe df_empresas
for i, column in enumerate(column_names):
    df_empresas.loc[problematic_rows.index, column] = split_data[i]
```

```python
# Ahora se deben verificar las filas corregidas
df_empresas.loc[problematic_rows.index].head()
```

```python
# Como no se pudieron corregir otdas, se deben eliminar filas que no pudieron ser corregidas
df_empresas.dropna(subset=['NIT'], inplace=True)
```

```python
# Ahora se debe verificar si quedan valores nulos o NaN en las columnas importantes
df_empresas.isnull().sum()
```

```python
df_empresas.head(40) # Revisamos una vez mas las 40 filas para ver cambios:
```

```python
df_empresas.dtypes
```

```python
# Si bien ya redistribuimos las observaciones en las filas, falta eliminar símbolos como $ y comas para poder hacer operaciones con columnas que tiene datos

columns_to_clean = ['INGRESOS OPERACIONALES', 'GANANCIA (PÉRDIDA)', 'TOTAL ACTIVOS', 'TOTAL PASIVOS', 'TOTAL PATRIMONIO']

for column in columns_to_clean:
    df_empresas[column] = df_empresas[column].replace({'\$': '', ',': ''}, regex=True)
```

```python
# Si bien sale un error de syntaxis, esto no impide seguir para convertir las columnas a float:
for column in columns_to_clean:
    df_empresas[column] = df_empresas[column].astype(float)
```

```python
# Ahora se verifica que las columnas sean de tipo float
print(df_empresas[columns_to_clean].dtypes)
```

```python
# Para asegurar la transformación y la limpieza, se convierte a float manejando posibles errores:

for column in columns_to_clean:
    df_empresas[column] = pd.to_numeric(df_empresas[column], errors='coerce')
```
