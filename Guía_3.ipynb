## Limpieza de los datos

```python
df_secop1.columns
```

```python
def Limpiar_texto (df, column_name):
    df[column_name] = df[column_name].str.strip() ##Limpia datos en blanco
    df[column_name] = df[column_name].str.upper() #coloca todo en mayúsculas
    df[column_name] = df[column_name].replace('  ',' ') #limpia dobles espacios por espacios simples
    df[column_name] = df[column_name].replace('$','') #reemplaza $ por espacio en blanco
    df[column_name] = df[column_name].replace(r'[^\w\d\s\.]', '', regex=True) #reemplaza caracteres especiales
    df[column_name] = df[column_name].str.replace('Á', 'A') ## limpieza de las tildes 
    df[column_name] = df[column_name].str.replace('É', 'E')
    df[column_name] = df[column_name].str.replace('Í', 'I')
    df[column_name] = df[column_name].str.replace('Ó', 'O')
    df[column_name] = df[column_name].str.replace('Ú', 'U')
    return df
```

```python
Limpiar_texto (df_secop1,'nombre_entidad')
Limpiar_texto (df_secop1,'nit_entidad')
Limpiar_texto (df_secop1,'direccion_entidad')
Limpiar_texto (df_secop1,'telefono_entidad')
Limpiar_texto (df_secop1,'mision_y_vision')
Limpiar_texto (df_secop1,'perspectiva_estrategica')
Limpiar_texto (df_secop1,'info_contacto')
Limpiar_texto (df_secop1,'ppto_total')
Limpiar_texto (df_secop1,'limite_menor_cuantia')
Limpiar_texto (df_secop1,'limite_minima_cuantia')
Limpiar_texto (df_secop1,'municipio_paa')
Limpiar_texto (df_secop1,'departamento_paa')
Limpiar_texto (df_secop1,'localidad_paa')
```

```python
#funcion para hacer limpieza en columnas numéricas
def Limpieza_Numericas (df,column_name):
    df[column_name] = df[column_name].replace('[^0-9]','',regex=True) #quitar todo lo diferente a números
    df[column_name] = df[column_name].replace('','0').astype(float) # colocarle un cero a las variables vacías, con el float, cambio el formato a la variable. 
    return df
```

```python
#limpieza de variables numéricas y transformación del tipo de dato.
Limpieza_Numericas (df_secop1,"nit_entidad")
Limpieza_Numericas (df_secop1,"telefono_entidad")
Limpieza_Numericas (df_secop1,"ppto_total")
Limpieza_Numericas (df_secop1,"limite_menor_cuantia")
Limpieza_Numericas (df_secop1,"limite_minima_cuantia")
Limpieza_Numericas (df_secop2,'nit_entidad')
Limpieza_Numericas (df_secop2,'telefono_contacto')
Limpieza_Numericas (df_secop2,'valor_presupuesto_general')
Limpieza_Numericas (df_secop2,'presupuesto_menor_cuantia')
Limpieza_Numericas (df_secop2,'presupuesto_minima_cuantia')
```

```python
def extraer_nueve_digitos(df, column_name):
    return df[column_name].astype(str).str[:9]
```

```python
df_secop = df_secop1.merge(df_secop2, on=['nit_entidad'], how='left')
```
